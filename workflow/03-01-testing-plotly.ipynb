```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
from sklearn.metrics import mean_squared_error
import os
import numpy as np
from plotly.io import to_html, write_html
```

```python
df = pd.read_csv('../modelOutputs/see-run04-regression_longstocks_100cells_40epochs-droptail.csv')
df['Close_t-1'] = df['Close'].shift(periods=1)
df['predClose_t-1'] = df['predClose'].shift(periods=1)
df.head()
```

```python
df['label'] = np.where(df['Close'] - df['Close_t-1'] > 0, 1, 0)
```

```python
df['predLabel'] = np.where(df['predClose'] - df['Close_t-1'] > 0, 1, 0)
df = df[1:]
```

```python
(df['predLabel'] == df['label']).value_counts()
```

```python
fig  = go.Figure()
fig.add_trace(go.Scatter(x=df['Date'], y=df['predClose'],
                    name='predicted price'))
fig.add_trace(go.Scattergl(x=df['Date'], y=df['Close'],
                    name='real price'))
#fig.add_trace(go.Scatter(x=df['Date'], y=df['Close_t-1'],
                    #name='real price yesterday'))
    # mode='markers', marker={'size':3}
```

```python
write_html(fig, './stock-name.html')
```

```python
df = df[1:]
error_DL = mean_squared_error(df['Close'], df['predClose'], squared=False)
error_priceYest = mean_squared_error(df['Close'], df['Close_t-1'], squared=False)
```

```python
print(error_DL, error_priceYest)
```
